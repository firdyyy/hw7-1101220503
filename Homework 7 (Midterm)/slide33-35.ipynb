```python
n = 100
np.random.seed(100)
```

```python
fea_lst = []
y_lst = []
```

```python
files = ['C:\\Users\\firdi\\Downloads\\Activity Recognition from Single Chest-Mounted Accelerometer\\Activity Recognition from Single Chest-Mounted Accelerometer\\1.csv', 'C:\\Users\\firdi\\Downloads\\Activity Recognition from Single Chest-Mounted Accelerometer\\Activity Recognition from Single Chest-Mounted Accelerometer\\2.csv', 'C:\\Users\\firdi\\Downloads\\Activity Recognition from Single Chest-Mounted Accelerometer\\Activity Recognition from Single Chest-Mounted Accelerometer\\3.csv', 'C:\\Users\\firdi\\Downloads\\Activity Recognition from Single Chest-Mounted Accelerometer\\Activity Recognition from Single Chest-Mounted Accelerometer\\4.csv', 'C:\\Users\\firdi\\Downloads\\Activity Recognition from Single Chest-Mounted Accelerometer\\Activity Recognition from Single Chest-Mounted Accelerometer\\5.csv', 'C:\\Users\\firdi\\Downloads\\Activity Recognition from Single Chest-Mounted Accelerometer\\Activity Recognition from Single Chest-Mounted Accelerometer\\6.csv', 'C:\\Users\\firdi\\Downloads\\Activity Recognition from Single Chest-Mounted Accelerometer\\Activity Recognition from Single Chest-Mounted Accelerometer\\7.csv', 'C:\\Users\\firdi\\Downloads\\Activity Recognition from Single Chest-Mounted Accelerometer\\Activity Recognition from Single Chest-Mounted Accelerometer\\8.csv']
```

```python
for i, f in enumerate(files):   
    print(i, f)
```

```python
df = pd.read_csv(f)
df.ffill(inplace=True)
df.bfill(inplace=True)
```

```python
data = df.values
```

```python
idx = np.random.choice(data.shape[0], min(n, data.shape[0]), replace=False)
arr = data[idx, :]
```

```python
fea_lst.append(arr)
y_lst.append(np.full(arr.shape[0], i))
```

```python
print(f'df.shape {df.shape}, pilih {arr.shape[0]} fitur')
```

```python
fea_name = list(df.columns)
x_data = np.vstack(fea_lst)
y_label = np.hstack(y_lst)
```

```python
print(y_label)
print(f'x_data shape: {x_data.shape}')
print(f'y_label shape: {y_label.shape}')
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=0)
clf.fit(x_data, y_label)
```

```python
print(f'Akurasi Model: {clf.score(x_data, y_label)}')
```

```python
fea_dict = {}
for i, fea in enumerate(fea_name):
    fea_dict[fea] = clf.feature_importances_[i]
```

```python
fea_import = [(v, k) for k, v in fea_dict.items()]
fea_import = sorted(fea_import, reverse=True)
```

```python
print(f'Pentingnya Fitur: {fea_import}')
```

```python
header = "Feature, Importance"
out_path = 'C:/Users/firdi/anaconda_projects/Hasilslide33-35'
```

```python
if not os.path.exists(out_path):
    os.makedirs(out_path)
```

```python
for i, (importance, feature) in enumerate(fea_import):
    file_path = os.path.join(out_path, f'feature_importance_{i}.csv')
    with open(file_path, 'w') as f:
        f.write(f"{header}\n")
        f.write(f"{feature}, {importance}\n")
```
